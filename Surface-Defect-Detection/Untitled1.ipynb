```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
from docx import Document
import io

# Set the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract'  # Update your path accordingly

def pdf_to_word(pdf_path, word_path, lang='vie'):
    doc = Document()
    pdf_document = fitz.open(pdf_path)

    # Loop through each page in the PDF
    for page_number in range(len(pdf_document)):
        page = pdf_document.load_page(page_number)

        # Render page to an image
        pix = page.get_pixmap(matrix=fitz.Matrix(300/72, 300/72))  # Increase the resolution
        img = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

        # Perform OCR using Tesseract
        text = pytesseract.image_to_string(img, lang=lang)
        doc.add_paragraph(text)

        # Optionally add a page break after each page if not the last page
        if page_number < len(pdf_document) - 1:
            doc.add_page_break()

    pdf_document.close()
    doc.save(word_path)
    print(f"Document saved as {word_path}")

# Example usage:
pdf_path = 'D:\\toan 1-10.pdf'
word_path = 'D:\\toan 1-10.docx'
pdf_to_word(pdf_path, word_path)

```
